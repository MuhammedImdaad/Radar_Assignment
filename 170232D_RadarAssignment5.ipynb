# Q2 - Expression of received signal when there are two isotropic antenna elements rotating

image.png

In the above image, two isotropic antennas are rotating at a frequency of $f_m$ on a radius R. As it was said not to assume the position vector to the two antennas from the target is parallel, we can not follow the procedure what was followed during the lecturers. 

$\therefore$ under the following assumptions, cosine rule will be applied.

Assumptions:
* $|A(t)| = 1$ for signals for subsequent calculations
* $ R << d$
* position vector to the two antennas from the target is not parallel

phase calculation,
* At a time t, assuming the rotation speed is ω, the phase rotation is $\omega t$
* Since the starting phase is θ
* phase = $\omega t - θ$

using cosine rule, 
\begin{align}
d_1 &= \sqrt{R^2+d^2-2dR\cos(\omega_m t-\theta)} \\
&= d\sqrt{1+\frac{R^2}{d^2}-2\frac{R}{d}\cos(\omega_m t-\theta)} \\
&= d\sqrt{1-2\frac{R}{d}\cos(\omega_m t-\theta)} \because d \gg R \implies \frac{R^2}{d^2} \approx 0 \\
 d_1 &= d-R\cos(\omega_m t-\theta) \because \sqrt{1-x} \approx 1-\frac{x}{2}\\
\text{similarly we can obtain}\\
d_2 & =d+R\cos(\omega_m t-\theta) 
\end{align}

Let's find each antenna target receiver signal, $V_{A1}$ and $V_{A2}$ \\
where $ \omega_m t = 2\pi f_m t$ 

\begin{align}
V_{A1} &=\sin(2\pi (f_c+f_0)t+\phi_1) \\
&=\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda} d_1) \\
&=\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \\
similarly,\\
V_{A2} &=\sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) \\
\end{align}

$\therefore$ Signal received by the target due to two rotating antennas, $V_{Rx}

\begin{align}
V_{Rx} &= V_{A1} + V_{A2} \\
V_{Rx} &= \sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \ + \sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) 
\end{align}

# Q3 - Signal generation of the combined carrier modulated with the message 

Now there is a third antenna stationed on the origin which carries a message modulated using the carrier amplitude. The end receive signal by the target would be the combination of these 3 antennas

## Origin signal received by the Target
\begin{align}
V_{A0}&=(1 + x_1(t) + x_2(t)\sin(ω_1 t) - M\sin(2\pi f_mt))\sin\left(2\pi f_ct - \frac{2\pi d}{\lambda}\right) \\
\text{Assuming voice and identity signal is negligible}\\
V_{A0}&=(1 - M\sin(2\pi f_mt))\sin\left(2\pi f_ct - \frac{2\pi d}{\lambda}\right)
\end{align}

## Composition
Received composite signal is the combination of $V_{A0}+V_{A1}+V_{A2}$.

\begin{align}
V_{R_x}&=V_{A0}+V_{A1}+V_{A2}\\
V_{R_x}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct - \frac{2\pi d}{\lambda}\right)+\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \\&+\sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) 
\end{align}

If we take $V_{A0}$ as our reference signal, the relative phase shift can be obtained as follows:
\begin{align}
\Delta\phi_1&= \frac{2\pi}{\lambda}(d-d_1) &=\frac{2\pi}{\lambda}R\cos(2\pi f_mt-\theta) &=\Delta\phi \\
\Delta\phi_2&= \frac{2\pi}{\lambda}(d2-d)&=-\frac{2\pi}{\lambda}R\cos(2\pi f_mt-\theta) &=-\Delta\phi \\
\end{align}

$\therefore$ 
\begin{align}
V_{R_x}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+\sin(2\pi (f_c+f_0)t+\Delta\phi) \\&+\sin(2\pi (f_c-f_0)t-\Delta\phi)\\
&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+2\sin(2\pi f_ct)\cos(2\pi f_0t +\Delta\phi )\\
&=\{1-M\sin(2\pi f_mt)+2\cos(2\pi f_0t +\Delta\phi )\}\sin\left(2\pi f_ct \right)\\
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_0t +\frac{2\pi R}{\lambda}\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right)\\
\text{let $\delta$ be the FM modulation index and it is equal to,} \\
δ &= \frac{2 \pi R}{λ}\\
\text{simpliefied received signal}\\
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_0t +\delta\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right)
\end{align}

## Implementation of the signal generation

```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert
```

```python
# Hypothetical Parameters 
f_c = 1e6 # Carrier Frequency 
f_o = 960 # sub carrier offset
f_m = 30 # Rotation frequency of VOR system
d = 200*1e3 # Distance to the target
theta = ((2*pi)/360)*30 # Phase angle from magnetic north 

# general paramteres
C = 3*1e8 # Speed of Light
M = 0.3 # AM Modulation index
delta = 16  # FM Modulation Index
```

```python
#Radius of the rotating antenna
lmbda = C / f_c # Wavelength
R = 8*lmbda/pi # Radius of the VOR antenna 
print('Radius of the VOR antenna circle (R):',str(R),'m')

#phase shift calculation
phase_shift = (2*pi*d)/lmbda
print('Phase shift due to distance: ',str(phase_shift),'rad')
```

```python
# angular frequencies
w_m = 2*pi*f_m
w_o = 2*pi*f_o
w_c = 2*pi*f_c

# To filter the constant value of the modulator
E = 1.0/M

# sampling frequency
F_s = int(10*f_c)

# time vector
t = np.linspace(0, 1, F_s+1)
```

```python
# derive the signals programmatically
d_1 = d - R*np.cos(w_m*t-theta)
d_2 = d + R*np.cos(w_m*t-theta)

delt_phi_1 = (2*pi/lmbda)*(d_1)
delt_phi_2 = (2*pi/lmbda)*(d_2)

V_A1 = E*np.sin((w_c+w_o)*t - delt_phi_1)
V_A2 = E*np.sin((w_c-w_o)*t - delt_phi_2)

V_A0 = E*(1-M*np.sin(w_m*t))*np.sin(w_c*t-phase_shift)

end_indice = 3000
plt.figure(figsize=(25,5))
plt.plot(t[0:end_indice], V_A1[0:end_indice])
plt.ylim((-4 ,4))
plt.title('Recieved Signal from Antenna Element A1')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(25,5))
plt.plot(t[0:end_indice], V_A2[0:end_indice])
plt.ylim((-4 ,4))
plt.title('Recieved Signal from Antenna Element A2')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(25,5))
plt.plot(t[:end_indice*300], V_A0[:end_indice*300])
plt.title('Recieved Signal from Antenna Element A0')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()
```

```python
V_Rx = V_A0+V_A1+V_A2 #Rx signal

plt.figure(figsize=(50,10))
plt.plot(t[0:1000000], V_Rx[0:1000000])
plt.title('Recieved composite signal $V_{Rx}$ ')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()
```

## Amplitude demodulation of the received signal

Using an envelope detector followed by a low pass filter would remove the carrier sinusoidal. 

\begin{align}
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_0t +\delta\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right) \times 2\sin(2\pi f_ct)\\
&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_0t +\delta \cos(2\pi f_mt-\theta) \right)\right\}(1-\cos(4\pi f_ct))\\
\text{applying a LPF centered on $f_c$}\\
&=1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_0t +\delta \cos(2\pi f_mt-\theta) \right)
\end{align}



```python
# Low pass filter
def LPF(cutoff,fs,N,plot = True ,name = ""):
  H = np.zeros((1,N))
  H[0,0:cutoff+1] = np.ones((1,cutoff+1))
  H[0,-cutoff:] = np.ones((1,cutoff))
  H=H[0]
  print(H.shape)

  if plot:
    plt.figure(figsize=(20,4))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency (Hz)')
    ax=plt.title('Lowpass filter for : '+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff+50))
    # plt.show()
  return H
```

```python
#envelope detector
V_mod = V_Rx*(2*np.sin(w_c*t-phase_shift))
mod_FFT=np.fft.fft(V_mod)
mod_abs_FFT = (np.abs(mod_FFT)**1).flatten()

# #low pass filter to obtain baseband signal with a BW of 10KHz
H0_LPF=LPF(10000,F_s,len(V_mod),0,'filter product detector output') 
plot_FFT(mod_abs_FFT,F_s,(-2.7e6,2.7e6),'Output of the product detector $V_{Rx}$')
plt.plot(freq_axis,H0_LPF,'r')
plt.show()
```

```python
# applying LPF 
W_mod = mod_FFT*H0_LPF

V_demod = np.real(np.fft.ifft(W_mod))

plot_FFT((np.abs(W_mod)**1).flatten(),F_s,(-2000,2000),'LPF filtered AM demodulated $V_{Rx}$')                            

#plot time domain demodulated signal
fig= plt.figure(figsize=(50,10))
plt.plot(t, V_demod)
plt.title('Demodulated Signal')
plt.xlabel('Time (s)') 
plt.ylabel('Magnitude')
plt.xlim((0, 0.08))
plt.show()
```

## Extract the variable signal

As same as the reference signal, now with a Band-Pass filter centered around 960Hz we can extract the variable signal and obtain the $2\cos\left(2\pi f_0t +\delta \cos(2\pi f_mt-\theta) \right)$ signal .

```python
def BPF(cutoff_L,cutoff_R,fs,N,plot,name):
  L_idx = cutoff_L
  R_idx = cutoff_R

  H = np.zeros((1,N))
  print(H.shape)
  H[0,cutoff_L:cutoff_R+1 ]= np.ones((1,cutoff_R-cutoff_L+1))
  H[0,-cutoff_R-1:-cutoff_L]=np.ones((1,cutoff_R-cutoff_L+1))
  H = H[0]
  print(H.shape)

  if plot:
    plt.figure(figsize=(20,4))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency (Hz)')
    ax=plt.title('Banpass filter for : '+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff_R+50))
  return H   
```

```python
#apply the BPF
H_BPF=BPF(f_o-700,f_o+700,F_s,len(abs_Rx_FFT),0,'960Hz variable reference signal')
W_filt_var = W_mod*H_BPF

filt_var_ref_comp = (np.fft.ifft(W_filt_var))
filt_var_ref = np.real(filt_var_ref_comp)

plt.figure(figsize=(15,5))
plt.plot(freq_axis, 2.0/N * np.abs(W_mod),"r")
plt.plot(freq_axis,H_BPF,"g")
plt.xlabel('Frequency (Hz)')
plt.ylabel('$|H(f)|^2$')
plt.title('Filtering variable Reference 960Hz signal using Bnad pass Filter')
plt.xlim((-2060,2060))
plt.show()
```

## FM demodulate the variable signal

$2\cos\left(2\pi f_0t +\delta \cos(2\pi f_mt-\theta) \right)$.

Once the variable signal goes through the differential phase detection process the FM demodulation can be obtained as follows.

$Θ(t) = \delta \cos(2\pi f_mt-\theta)$ 

$ \frac{dΘ(t)}{dt} = -\delta ω_m sin(2\pi f_mt-\theta)$

this detection process will continue,

```python
freq_dev=30*16
# Compute the analytic signal, using the Hilbert transform.
R_x = hilbert(filt_var_ref)*np.exp(-1j*2*np.pi*f_o*t)
phi = np.arctan2(np.imag(R_x), np.real(R_x))
fitered_var = np.diff(np.unwrap(phi)/(2*pi*freq_dev))*t.shape[0]
fitered_var=(fitered_var-np.mean(fitered_var))

fig= plt.figure(figsize=(20,4))
plt.plot(t[:-1], np.real(fitered_var), 'r',label="Demodulated")
plt.xlim((0,0.15))
plt.show()
```

## determine the phase difference

In the VOR receiver, now the two signals (reference signal and variable signal) should be passed through a phase comparator to determine the phase difference between two signals. 

```python
ref_fft = np.fft.fft(filtered_ref - np.mean(filtered_ref))
var_fft = np.fft.fft(fitered_var- np.mean(fitered_var))

#get the w_m frequency bin
var_fft_bin= np.argmax(abs(var_fft))
ref_fft_bin = np.argmax(abs(ref_fft))

var_phase=np.angle(var_fft[var_fft_bin])
ref_phase=np.angle(ref_fft[ref_fft_bin])

phase_diff=np.abs(var_phase-ref_phase)

print('phase difference : ',np.round(phase_diff*180/pi,2),'deg')
```

This is the same as the initial Θ value for which the process started making calculations. Hence the process is accurate and in accordance to the fundermental theories

# Q5 - Electronical rotation of circular antenna array

According to the previous calculations, the cirular antenna pair needs to be rotating at a 30Hz frequency on a radius of 763.94m. This is equivalent to 1800rpm. The net gravitational force needs at the origin is significant and mechanical setup would be incapable/impractical to handle such force. 

Therefore electronically switching a circular array would be the ideal solution. For this, fixed number of omni directional antennas are arranged in the circle with the radius $R$ of previous setup. Rotation can be simulated by activating/deactivating radial pairs of the antenna setup. 24 pairs of such antennas are used for the implementation. 

However the switching on/off these antennas with the given high input power, there will be induced RF radiations which can cause unwanted interference in the target. Therefore radiation cannot be switched abruptly. To address this, zero switching concept has been introduced and the technique is called cosine blending function which will be used for ramping up/down the power.

We can divide antenna pairs into odd and even groups and handing over will be done from odd-to-odd and even-to-even. Thus there will be 12 each such odd/even pairs in the commercial DVOR system. 


calculating the blending frequency, 

\begin{align}
f_{blend}&=\frac{2\pi f_m}{4\times(2\pi / \text{antenna count})}\\
&=\frac{2\pi \times 30}{4\times(2\pi /48)}\\
&=360Hz
\end{align}

Blending odd and even function generation

\begin{align}
\text{cosine blending function for odd antennas}&= \sin(2\pi f_{blend}t)\\
\text{cosine blending function for even antennas}&= \cos(2\pi f_{blend}t)
\end{align}

```python
antennas=48
blend_freq= (np.pi*f_m)/(2*(2*np.pi / antennas))
pulse_width= 2/(f_m*antennas)
print("cosine blending frequency: ", blend_freq)
```

creation of the odd and even functions to simulate the rotational array of antenna switching

```python
halfw_sampled=int(0.5*pulse_width*F_s)+1
sample_period=int(F_s/f_m)+1
rep=31
beta=360/antennas
t_arr=np.arange(0,2/f_m,1/F_s)

discr_sin=np.sin(blend_freq*t_arr)
discr_sin[2*halfw_sampled-1:sample_period]=0
odd_func=np.array(list(discr_sin[:sample_period])*rep)

discr_cos=np.cos(blend_freq*t_arr)
discr_cos[halfw_sampled-1:sample_period-halfw_sampled]=0
even_func=np.array(list(discr_cos[:sample_period])*rep)
```

```python
#simulate rotation
V_Rx=0
for i in range(antennas):
  if i==0:
    angle=beta/2
    pulse=even_func
  else:
    angle+=beta
    pulse=np.array(list(np.zeros(((i-1)*halfw_sampled)))+list(odd_func))
  blend=pulse[0:N]
  dx=R*np.cos((theta-angle)*np.pi/180)
  phi_1=2*np.pi*(d-dx)/lmbda
  phi_2=2*np.pi*(d+dx)/lmbda

  upper_sp=E*np.sin(2*np.pi*(f_c+f_0)*t-phi_1)*blend
  lower_sp=E*np.sin(2*np.pi*(f_c-f_0)*t-phi_2)*blend
  V_Rx+=upper_sp+lower_sp
```

```python
plt.figure(figsize=(50,10))
plt.plot(t[0:1000000], (V_A0+V_Rx)[0:1000000])
plt.title('composite signal $V_{Rx}$')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()
```

```python
elect_fft = (np.fft.fft(E*(1-M*np.sin(w_m*t))*np.sin(w_c*t-phase_shift)+V_Rx))

plt.figure(figsize=(15,5))
plt.plot(freq_axis, 2.0/N * np.abs(elect_fft),"r")
plt.xlabel('Frequency (Hz)')
plt.ylabel('$|H(f)|^2$')
plt.title('Frequency response of Simulated Antenna Array')
plt.xlim((1e6-2000,1e6+2000))
plt.show()

plt.figure(figsize=(15,5))
plt.plot(freq_axis, 2.0/N * np.abs(Rx_FFT),"r")
plt.xlabel('Frequency (Hz)')
plt.ylabel('$|H(f)|^2$')
plt.title('Frequency response of Actual Antenna Array')
plt.xlim((1e6-2000,1e6+2000))
plt.show()
```

Hence with the electronically stimulated antenna array, the target which is possibly a VOR receiver in a plane would hardly find any difference compared to a theoratical mechanical pair of antennas. This is due to the cosine blending. 

However due to physical behaviour of the commercial setups, zero-switching would not be perfect thus induce some minor RF interruptions.
